```python
#defining a function to remove unrequired asswmb

def remove_char(tweet):
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet)
    tweet = re.sub(r'#', '', tweet)
    tweet = re.sub(r'[0-9]', '', tweet)
    
    return tweet
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vanderSentimentAnalyzer = SentimentIntensityAnalyzer()
```

```python
def getVanderScore(tweet):    
    vs = vanderSentimentAnalyzer.polarity_scores(tweet)
    score = vs['compound']
    return score

def getVanderSentiment(score):    
    if (score >= 0.05): 
        return "Positive"
    
    elif (score < 0.05 and score > -0.05):
        return "Neutral"
    
    elif (score <= -0.05):    
        return "Negative"
    
    return score
```

```python
dff2['vander_score'] = dff2['body'].apply(lambda tweet: getVanderScore(tweet))
dff2['vander_sentiment'] = dff2['vander_score'].apply(lambda score: getVanderSentiment(score))
dff2[['tickers_symbol','post_date','vander_score', 'vander_sentiment']].sample(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= dff2[['vander_score']]
y=dff2[['vander_sentiment']]
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=47, 
                                   test_size=0.25, 
                                   shuffle=True)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred= lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred))
```
